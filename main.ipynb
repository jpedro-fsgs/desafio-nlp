```python
import pydantic
```

```python
import json

with open("dados_grupo_estudos/biblioteca_aneel_gov_br_legislacao_2016_metadados.json", "r") as f:
	dados_2016 = json.load(f)

with open("dados_grupo_estudos/biblioteca_aneel_gov_br_legislacao_2021_metadados.json", "r") as f:
	dados_2021 = json.load(f)

with open("dados_grupo_estudos/biblioteca_aneel_gov_br_legislacao_2022_metadados.json", "r") as f:
	dados_2022 = json.load(f)
	
(len(dados_2016), len(dados_2021), len(dados_2022))
```

```python
from model_dados import Model

try:
    dados_2016 = Model.model_validate(dados_2016, strict=True)
    dados_2021 = Model.model_validate(dados_2021, strict=True)
    dados_2022 = Model.model_validate(dados_2022, strict=True)
except pydantic.ValidationError as e:
    print("Validation error:", e)
```

```python
for k,v in list(dados_2016.root.items()) + list(dados_2021.root.items()) + list(dados_2022.root.items()):
    for registro in v.registros:
        for pdf in registro.pdfs:
            pdf.baixado = False
```

```python
with open("dados_grupo_estudos_new/biblioteca_aneel_gov_br_legislacao_2016_metadados.json", "w") as f:
    json.dump(dados_2016.model_dump(), f, indent=4)

with open("dados_grupo_estudos_new/biblioteca_aneel_gov_br_legislacao_2021_metadados.json", "w") as f:
    json.dump(dados_2021.model_dump(), f, indent=4)

with open("dados_grupo_estudos_new/biblioteca_aneel_gov_br_legislacao_2022_metadados.json", "w") as f:  
    json.dump(dados_2022.model_dump(), f, indent=4)
```
